```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('train.csv')

dados.head(10)
```

```python
notas = dados[['SG_UF_RESIDENCIA','NU_IDADE','TP_SEXO','TP_COR_RACA','TP_NACIONALIDADE','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU','TP_ESCOLA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO','NU_NOTA_MT']]

notas = notas.dropna(axis=0, how='all')
notas = notas.fillna(0)
```

```python
notas.head(10)
```

```python
notas.corr().round(4)
```

```python
fig,ax  = plt.subplots(figsize =(20,6))
ax = notas['NU_NOTA_MT'].plot(fontsize=14)
```

```python
ax = sns.distplot(notas['NU_NOTA_MT'])
```

```python
ax = sns.pairplot(notas,y_vars='NU_NOTA_MT',x_vars=['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO'],kind='reg')
```

```python
ax = sns.jointplot(x="NU_NOTA_CN",y="NU_NOTA_MT",data=notas,kind='reg')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

y= notas['NU_NOTA_MT']
x= notas[['NU_IDADE','TP_COR_RACA','TP_NACIONALIDADE','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU','TP_ESCOLA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO']]
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
print('R² = {}'.format(modelo.score(x,y).round(2)))
```

```python
test = pd.read_csv('test.csv')
```

```python
x_test = test[['NU_IDADE','TP_COR_RACA','TP_NACIONALIDADE','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU','TP_ESCOLA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO']]
x_test = x_test.dropna(axis=0, how='all')
x_test = x_test.fillna(0)
```

```python
modelo.predict(x_test).round(1)
```

```python
test['NU_NOTA_MT'] = modelo.predict(x_test).round(1)
```

```python
test.head(10)
```

```python
answer = test[['NU_INSCRICAO','NU_NOTA_MT']]
answer.head(10)
```
